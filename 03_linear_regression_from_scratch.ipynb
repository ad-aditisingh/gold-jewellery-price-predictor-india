```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("jewellery_price_india.csv")
df.head()
```

```python
X = df["weight_grams"].values
y = df["jewellery_price"].values
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y, alpha=0.5)
plt.xlabel("Weight (grams)")
plt.ylabel("Jewellery Price (â‚¹)")
plt.title("Weight vs Jewellery Price")
plt.show()
```

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# Initialize parameters
m = 0.0
c = 0.0

learning_rate = 0.000001
epochs = 100
n = len(X)

for epoch in range(epochs):
    y_pred = m * X + c

    dm = (-2/n) * np.sum(X * (y - y_pred))
    dc = (-2/n) * np.sum(y - y_pred)

    m = m - learning_rate * dm
    c = c - learning_rate * dc

    if epoch % 10 == 0:
        loss = mean_squared_error(y, y_pred)
        print(f"Epoch {epoch}: Loss={loss:.2f}")
```

```python
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y, alpha=0.3, label="Actual")
plt.plot(X, m*X + c, color="red", label="Regression Line")
plt.xlabel("Weight (grams)")
plt.ylabel("Jewellery Price (â‚¹)")
plt.legend()
plt.show()
```

Linear Regression from Scratch

Implemented linear regression using gradient descent

Used weight as the single feature

Learned slope and intercept without ML libraries

Model captures linear relationship between weight and price
